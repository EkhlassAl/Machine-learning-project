```python

class Car:
    def __init__(self, brand, model, year, color):
        self.brand = brand
        self.model = model
        self.year = year
        self.color = color

    def calculate_price(self):
        base_price = 20000  # السعر الأساسي للسيارة

        # تحديد قيم إضافية بناءً على المواصفات
        if self.year >= 2020:
            base_price += 5000
        else:
            base_price += 2000

        if self.color == 'أحمر':
            base_price += 1000

        return base_price


# استخدام المستخدم لإدخال مواصفات السيارة
brand = input("العلامة التجارية: ")
model = input("الموديل: ")
year = int(input("السنة: "))
color = input("اللون: ")

# إنشاء كائن Car بناءً على المواصفات المُدخلة
car = Car(brand, model, year, color)

# حساب سعر السيارة وطباعته
price = car.calculate_price()
print("سعر السيارة هو:", price)
```

```python
import numpy as np

class LogisticRegression:
    def __init__(self, learning_rate=0.001, num_iterations=500):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        num_samples, num_features = X.shape
        
        # تهيئة الوزن والانحدار الأولي
        self.weights = np.zeros(num_features)
        self.bias = 0
        
        # تحديث الوزن والانحدار باستخدام الانحدار العكسي
        for _ in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self._sigmoid(linear_model)
            
            dw = (1 / num_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / num_samples) * np.sum(y_predicted - y)
            
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
    
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self._sigmoid(linear_model)
        y_predicted_cls = [1 if i > 0.5 else 0 for i in y_predicted]
        return y_predicted_cls
    
    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))


#python
from sklearn.model_selection import train_test_split
from sklearn import datasets

# تحميل مجموعة بيانات السيارات
cancer = datasets.load_breast_cancer()
X, y = cancer.data, cancer.target

# تقسيم البيانات إلى مجموعة تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

# تدريب النموذج
regression = LogisticRegression(learning_rate=0.1, num_iterations=1000)
regression.fit(X_train, y_train)

# تقييم النموذج
y_pred = regression.predict(X_test)
accuracy = np.sum(y_pred == y_test) / len(y_test)
print("Accuracy:", accuracy)

```
